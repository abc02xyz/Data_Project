## 1. simple line chart

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data for plotting
t=np.arange(0.0,2.0,0.01)
s=1+np.sin(2*np.pi*t)
```

```python
fig,ax=plt.subplots()
ax.plot(t,s)
ax.set(xlabel='X',ylabel='Y',title='Simple Line Plot')
ax.grid()
```

##  2. subplot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create Data
x1=np.linspace(0.0,5.0)
x2=np.linspace(0.0,2.0)
```

```python
y1=np.cos(2*np.pi*x1)*np.exp(-x1)
y2=np.cos(2*np.pi*x2)
```

```python
# plt.subplot(2,1,1) space divide into 2 row and 1 column and the last 1 means the order of the arrangement
plt.subplot(2,1,1)
plt.plot(x1,y1,'o-')
plt.title('subplot 1')
plt.ylabel('oscillation_1')

#second plot
plt.subplot(2,1,2)
plt.plot(x2,y2,'.-')
plt.title('subplot 2')
plt.ylabel('oscillation_2')


```

## 3. Scatter plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
N=50
x=np.random.rand(N)
y=np.random.rand(N)
colors=np.random.rand(N)
area=np.pi*(15*np.random.rand(N))**2
```

```python
plt.scatter(x,y,c=colors,s=area,alpha=0.5)
```

## 4. Histogram

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
```

```python
# Generate Data -> normal Distribution
mu=1000
sigma=150
x=mu+sigma+np.random.randn(500)
# N(0,1)-> N(mu,sigma^2)
```

```python
num_bins=50
```

```python
fig,ax=plt.subplots()
n,bins,patches=ax.hist(x,num_bins,normed=1)
y=mlab.normpdf(bins,mu,sigma)
ax.plot(bins,y,'--')
ax.set_xlabel('X')
ax.set_ylabel('Probability Density')
ax.set_title(r'Histogram : $\mu=1000$,$\sigma=150$')
```
